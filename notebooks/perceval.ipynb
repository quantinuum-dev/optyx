```python
import sys
import numpy as np

sys.path.append('../')

from optyx import *
from discopy.monoidal import PRO
```

```python
import perceval as pcvl
import perceval.components as comp
```

```python
theta = np.pi / 4
phi = 0

r = np.exp(1j * phi) * np.sin(theta)
t = np.cos(theta)

optyx_bs = Split() @ Split() >> Id(PRO(1)) @ SWAP @ Id(PRO(1)) >> Scale(r) @ Scale(t) @ Scale(np.conj(t)) @ Scale(-np.conj(r)) >> Merge() @ Merge()

optyx_bs.draw()
```

```python
C1 = optyx_bs.to_perceval()
pcvl.pdisplay(C1)
```

```python
print(optyx_bs.prob(1))

analyzer = optyx_bs.prob_with_perceval(n_photons=1)
pcvl.pdisplay(analyzer)
```

```python
C = optyx_bs >> Scale(2) @ 1 >> optyx_bs

print(C.to_path())
C.draw()
```

```python
perc = C.to_perceval()

pcvl.pdisplay(perc)
```

```python
print(C.prob(1))

analyzer = C.prob_with_perceval(n_photons=1)
pcvl.pdisplay(analyzer)
```
